```python
def minmax(data):
    mx=mn=data[0]
    for i in data:
        if mx<i : mx=i
        if mn>i : mn=i
    return (mn,mx)
```

```python
def rev(data):
    for i in range(len(data)//2):
        data[i], data[len(data)-i-1] = data[len(data)-i-1], data[i]
    return data
```

```python
def unique(data):
    t=set(data)
    if len(data) == len(t): return True
    else: return False
```

```python
def read(path):
    f=open(path)
    lines=[]
    while True:
        try:
            line=f.readline()
            if line == '' : raise EOFError
            lines.append(line[:-1])
        except EOFError:
            for i in range(len(lines)):
                print(lines[len(lines)-i-1])
            return
```

```python
def factors(n):
    k=1
    rev=[]
    while k*k<n:
        if n%k==0:
            yield k
            rev.append(n//k)
        k+=1
        if k*k==n:
            yield k
    for factor in reversed(rev):
        yield factor
```

```python
def change(given, charged):
    bills = [50000, 10000, 5000, 1000, 500, 100, 50, 10, 5, 1]
    changed={}
    rest=given-charged
    for bill in bills:
        num, rest = rest//bill, rest%bill
        changed[bill]=num
    return changed
```

```python
def calculator():
    l=r=op=None
    while  True:
        s=input()
        if s=='Exit':
            return
        elif s=='Clr':
            l=r=op=None
        elif l == None:
            l = float(s)
        elif op == None:
            op = s
        elif r == None:
            r = float(s)
            if op=='+': print(l,op,r,'=',l+r)
            elif op=='-': print(l,op,r,'=',l-r)
            elif op=='*': print(l,op,r,'=',l*r)
            elif op=='/': print(l,op,r,'=',l/r)
            else: print('WorngInput')
```

```python
import random
def paradox(n):
    T=1000
    cnt=0
    for i in range(T):
        days=[random.randrange(365) for _ in range(n)]
        if len(set(days)) != len(days): cnt+=1
    return cnt/T

for i in range(101):
    print('Paradox probability for',i,'people is',paradox(i))
```
